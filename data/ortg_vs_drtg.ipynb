```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("rawDataProcessing/combined/99-23NBA_per_100_poss.csv")
df

df_starters = df[df["GS"] >41 ]

df_starters
```

```python
top_teams = df_starters[df_starters["Team Performance (20)"] == "top"]
bottom_teams = df_starters[df_starters["Team Performance (20)"] == "bottom"]


plt.figure(figsize=(10, 6))

#scatter plot for ortg and dtg for top teams and bottom teams
plt.scatter(top_teams['ORtg'], top_teams['DRtg'], color='green', label='Top Teams', alpha=0.5)
plt.scatter(bottom_teams['ORtg'], bottom_teams['DRtg'], color='red', label='Bottom Teams', alpha=0.5)

plt.title('Offensive vs. Defensive Ratings for Starters')
plt.xlabel('Offensive Rating (ORtg)')
plt.ylabel('Defensive Rating (DRtg)')
plt.legend()
plt.savefig(f"ORtg_vs_DRtg_starters", bbox_inches='tight')  
plt.show()

```
