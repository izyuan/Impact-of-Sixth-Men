```python
import pandas as pd

combined_df = pd.read_csv('combined/99-23NBA_per_game.csv')
combined_df2 = pd.read_csv('combined/99-23NBA_per_100_poss.csv')
```

```python
#filtering out to only bench players 
bench_players = combined_df[combined_df['GS'] < 41]
bench_players2 = combined_df2[combined_df2["GS"] <41 ]
#dictionary for top teams, average teams, and bottom teams
team_performance = {
    "1999-2000": {
        "top": ["LAL", "POR","IND", "UTH"],
        "bottom": ["CHI","VAN" , "GSW", "LAC"]
    },
    "2000-2001": {
        "top": ["LAL", "PHI","SAC", "SAS"],
        "bottom": ["CHI","VAN" , "GSW", "WAS"]
    },
    "2001-2002": {
        "top": ["SAC", "LAL","DAL", "SAS"],
        "bottom": ["CHI","DEN" , "GSW", "MEM"]
    },
    "2002-2003": {
        "top": ["MIN", "SAC","DAL", "SAS"],
        "bottom": ["MIA","TOR" , "CLE", "DEN"]
    },
    "2003-2004": {
        "top": ["IND", "MIN","LAL", "SAS"],
        "bottom": ["ORL","CHI" , "WAS", "LAC"]
    },
    "2004-2005": {
        "top": ["MIA", "PHO","SAS", "DAL"],
        "bottom": ["CHA","ATL" , "UTH", "NOH"]
    },
    "2005-2006": {
        "top": ["DET", "SAS", "DAL", "PHO"],
        "bottom": ["POR","NYK" , "CHA", "ATL"]
    },
    "2006-2007": {
        "top": ["DAL", "PHO","SAS", "DET"],
        "bottom": ["MEM","BOS" , "MIL", "ATL"]
    },
    "2007-2008": {
        "top": ["BOS", "DET","LAL", "NOH"],
        "bottom": ["MEM","MIA" , "SEA", "MIN"]
    },
    "2008-2009": {
        "top": ["CLE", "LAL","BOS", "ORL"],
        "bottom": ["SAC","WAS" , "LAC", "OKC"]
    },
    "2009-2010": {
        "top": ["CLE", "ORL","LAL", "DAL"],
        "bottom": ["NJN","MIN", "SAC", "WAS"]
    },
    "2010-2011": {
        "top": ["CHI", "SAS","MIA", "DAL"],
        "bottom": ["MIN","CLE" , "TOR", "WAS"]
    },
    "2011-2012": {
        "top": ["CHI", "SAS", "OKC", "MIA"],
        "bottom": ["CHA", "WAS", "NOH", "CLE"]
    },
    "2012-2013": {
        "top": ["MIA", "OKC", "SAS", "DEN"],
        "bottom": ["CHA", "ORL", "CLE", "PHO"]
    },
    "2013-2014": {
        "top": ["SAS", "LAC", "OKC", "IND"],
        "bottom": ["MIL", "PHI", "ORL", "UTH"]
    },
    "2014-2015": {
        "top": ["GSW", "ATL", "HOU", "LAC"],
        "bottom": ["MIN", "NYK", "PHI", "LAL"]
    },
    "2015-2016": {
        "top": ["GSW", "SAS", "CLE", "TOR"],
        "bottom": ["PHI", "LAL", "BRK", "PHO"]
    },
    "2016-2017": {
        "top": ["GSW", "SAS", "HOU", "BOS"],
        "bottom": ["BRK", "PHO", "LAL", "PHI"]
    },
    "2017-2018": {
        "top": ["HOU", "TOR", "GSW", "BOS"],
        "bottom": ["PHO", "MEM", "DAL", "ATL"]
    },
    "2018-2019": {
        "top": ["MIL", "TOR", "GSW", "DEN"],
        "bottom": ["NYK", "PHO", "CLE", "CHI"]
    },
    "2019-2020": {
        "top": ["MIL", "TOR", "LAL", "LAC"],
        "bottom": ["GSW", "MIN", "CLE", "DET"]
    },
    "2020-2021": {
        "top": ["UTH", "PHO", "PHI", "BRK"],
        "bottom": ["HOU", "DET", "ORL", "OKC"]
    },
    "2021-2022": {
        "top": ["PHO", "MEM", "GSW", "MIA"],
        "bottom": ["HOU", "ORL", "DET", "OKC"]
    },
    "2022-2023": {
        "top": ["MIL", "BOS", "PHI", "DEN"],
        "bottom": ["DET", "SAS", "HOU", "CHA"]
    }
}

def categorize_performance(row): #function to find the top, average, and bottom teams for that respective year 
    season = row['Season']
    team = row['Tm']
    for performance, teams in team_performance.get(season, {}).items():
        if team in teams:
            return performance
    return 'Average' 

def get_top_two(group): #function to return only the top two, this is used to find the top two bench players of each team 
    return group.iloc[:2]

bench_players['Team Performance'] = bench_players.apply(categorize_performance, axis=1)
bench_players2['Team Performance'] = bench_players2.apply(categorize_performance, axis=1)

sixthman = bench_players.sort_values(['Tm', 'Season', 'MP'], ascending=[True, True, False]) #grouping the teams by team, season, and match played to get the top 2 "sixth men"
sixthman2 = bench_players2.sort_values(['Tm', 'Season', 'MP'], ascending=[True, True, False])

top_two_per_team_season = sixthman.groupby(['Tm', 'Season']).apply(get_top_two)
top_two_per_team_season = top_two_per_team_season[top_two_per_team_season["Tm"] != "TOT"]
top_two_per_team_season = top_two_per_team_season.reset_index(drop=True)

top_two_per_team_season2 = sixthman2.groupby(['Tm', 'Season']).apply(get_top_two)
top_two_per_team_season2 = top_two_per_team_season2[top_two_per_team_season2["Tm"] != "TOT"]
top_two_per_team_season2 = top_two_per_team_season2.reset_index(drop=True)


top_two_per_team_season2["MPG"] = top_two_per_team_season["MP"]
top_two_per_team_season.to_csv("combined/top2SixthMenPerGame.csv")

```
