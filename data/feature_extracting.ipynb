```python
import pandas as pd

team_data = pd.read_csv("rawDataProcessing/combined/teamStatsPer100.csv")
ind_data = pd.read_csv("rawDataProcessing/combined/top2SixthMenPer100Poss.csv")

pd.set_option('display.max_columns', 999)
pd.set_option('display.max_rows', 999)
```

```python
"""needed game stats
team:
PTS, FG, FT, FGA, FTA, DRB, ORB, AST, STL, BLK, PF, TOV
individual: 
PTS+ FG +  FT - FGA - FTA + DRB, ORB + AST + STL + BLK/2 - PF -TO """

team_stat_columns = ['PTS', 'FG', 'FT', 'FGA', 'FTA', 'DRB', 'ORB', 'AST', 'STL', 'BLK', 'PF', 'TOV']
team_data.rename(columns={col: f'Team{col}' for col in team_stat_columns}, inplace=True)

if "Team" in team_data.columns:
    columns_to_drop = ['Unnamed: 0', 'Team', 'G', 'MP', 'TRB', 'FT%', '2P%', '2PA', '2P', '3P%', '3PA', '3P', 'FG%']
    team_data.drop(columns=columns_to_drop, inplace=True)

teamAndPlayerData = pd.merge(ind_data, team_data, on=['Tm', 'Season'], how='left')
print(teamAndPlayerData.columns.to_list())
teamAndPlayerData
```

```python
has_nan = teamAndPlayerData.isnull().values.any()
print(has_nan)
nan_counts = teamAndPlayerData.isnull().sum()
print(nan_counts)

columns_with_nan = teamAndPlayerData.columns[teamAndPlayerData.isnull().any()].tolist()
print(columns_with_nan)

rows_with_nan = teamAndPlayerData[teamAndPlayerData.isnull().any(axis=1)]
print(rows_with_nan)
```

it looks like the only missing data is 3 P% which is NaN because the player didn't shoot any threes probably. 
This is fine as we don't need three point percentage for the PIE 

Now it's time to calculate a composite score.

```python
player_numerators = (
    teamAndPlayerData['PTS'] + teamAndPlayerData['FG'] + teamAndPlayerData['FT'] - teamAndPlayerData['FGA'] - teamAndPlayerData['FTA']
    + teamAndPlayerData['DRB'] + teamAndPlayerData['ORB'] / 2
    + teamAndPlayerData['AST'] + teamAndPlayerData['STL']
    + teamAndPlayerData['BLK'] / 2 - teamAndPlayerData['PF'] - teamAndPlayerData['TOV']
)

game_denominators = (
    teamAndPlayerData['TeamPTS'] + teamAndPlayerData['TeamFG'] + teamAndPlayerData['TeamFT']
    - teamAndPlayerData['TeamFGA'] - teamAndPlayerData['TeamFTA']
    + teamAndPlayerData['TeamDRB'] + teamAndPlayerData['TeamORB'] / 2
    + teamAndPlayerData['TeamAST'] + teamAndPlayerData['TeamSTL']
    + teamAndPlayerData['TeamBLK'] / 2 - teamAndPlayerData['TeamPF'] - teamAndPlayerData['TeamTOV']
)

teamAndPlayerData["Composite"] = player_numerators/ game_denominators

teamAndPlayerData

```

```python
def get_top_two(group): #function to return only the top two, this is used to find the top two bench players of each team 
    return group.iloc[:2]

copy_df = teamAndPlayerData.copy()

max_mp = copy_df['MP'].max()
copy_df['NormalizedMP'] = copy_df['MP'] / max_mp

# Create a weighted combined metric: 70% NormalizedMP + 30% Composite
copy_df['WeightedCombinedMetric'] = 0.7 * copy_df['NormalizedMP'] + 0.3 * copy_df['Composite']
copy_df = copy_df.sort_values(['Tm', 'Season', 'WeightedCombinedMetric'], ascending=[True, True, False])


copy_df = copy_df.groupby(['Tm', 'Season']).apply(get_top_two)
copy_df = copy_df[copy_df["Tm"] != "TOT"]
copy_df = copy_df.reset_index(drop=True)
columns_to_keep = ['Player', 'Tm', 'Pos', 'Age', 'MP', 'Team Performance', 'Season', 'Composite']

analysis_dataframe = copy_df[columns_to_keep]

analysis_dataframe

analysis_dataframe.to_csv("analysis_dataframe2.csv")
```
